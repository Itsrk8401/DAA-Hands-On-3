```python
import numpy as np
import time
import matplotlib.pyplot as plt

def f(n):
    x = 1
    for i in range(1, n + 1):
        for j in range(1, n + 1):
            x = x + 1
    return x

# Timing and plotting
n_values = list(range(1, 35))  # You can adjust the range of n-values as needed
execution_times = []

for n in n_values:
    start_time = time.time()
    f(n)
    end_time = time.time()
    execution_times.append(end_time - start_time)

# Plotting
plt.plot(n_values, execution_times, 'o', label='Actual Data')

# Fit a quadratic curve to the data
coefficients = np.polyfit(n_values, execution_times, 2)
fitted_curve = np.polyval(coefficients, n_values)
plt.plot(n_values, fitted_curve, label='Quadratic Fit', linestyle='dashed')

plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.legend()
plt.show()
```
